```python
import json

def load_jsonl(file_path):
    data = []
    with open(file_path, 'r', encoding='utf-8') as f:
        for line in f:
            data.append(json.loads(line))
    return data

# Указываем абсолютные пути к обоим файлам
train_path = r'C:\Users\Пользователь\Desktop\Project_21\data\train.jsonl'
test_path = r'C:\Users\Пользователь\Desktop\Project_21\data\test.jsonl'

try:
    train_data = load_jsonl(train_path)
    test_data = load_jsonl(test_path)
    
    print(f"✅ В файле TRAIN загружено: {len(train_data)} примеров (Ожидается 90)")
    print(f"✅ В файле TEST загружено: {len(test_data)} примеров (Ожидается 20)")
    
    # Проверка на Data Leakage (чтобы примеры из теста не дублировались в обучении)
    train_inputs = set([item['input'] for item in train_data])
    leakage = 0
    for item in test_data:
        if item['input'] in train_inputs:
            leakage += 1
            
    if leakage == 0:
        print("✅ Утечек данных (Data Leakage) не обнаружено! Тестовые вопросы уникальны.")
    else:
        print(f"⚠️ Внимание: {leakage} примеров из теста дублируются в обучении.")

except FileNotFoundError as e:
    print(f"❌ Ошибка: Не найден файл {e.filename}")
```
